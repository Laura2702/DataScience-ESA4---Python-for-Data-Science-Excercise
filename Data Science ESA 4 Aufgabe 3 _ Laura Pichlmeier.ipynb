## 12. Create a new data frame from the original where the area is changed as follows: all countries with > 1000000 get BIG and <= 1000000 get SMALL in the cell replaced!

```python
#all countries with > 1000000 get BIG and <= 1000000 get SMALL in the cell replaced!
df3 = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/countries.csv')
df3['Area'] = np.where(df3['Area'] > 1000000, 'BIG','SMALL')
df3
```
